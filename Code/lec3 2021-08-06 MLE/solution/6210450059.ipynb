```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import scipy.stats as stats
import math
from matplotlib import pyplot as plt
```

```python
class GuassianMLE:
    
    def __init__(self):
        self.mu =0
        self.sigma=0
    def fit(self,x):
        self.mu=np.average(x)
        self.sigma=np.std(x)
```

```python
def gaussain_mackarels(x):
    fish = pd.read_csv('fish_dataset.csv',names=['fish_length','class'])
    fish_Mackerels=fish.loc[fish['class']==1] #Mackerels=1
    x_Mackerels=np.array(fish_Mackerels['fish_length'])
    h_Mackarels=GuassianMLE()
    h_Mackarels.fit(x_Mackerels)
    return stats.norm.pdf(x, h_Mackarels.mu, h_Mackarels.sigma)
    
```

```python
def gaussain_salmons(x):
    fish = pd.read_csv('fish_dataset.csv',names=['fish_length','class'])
    fish_Salmons=fish.loc[fish['class']==2]#Salmons=2
    x_Salmons=np.array(fish_Salmons['fish_length'])
    h_Salmons=GuassianMLE()
    h_Salmons.fit(x_Salmons)
    return stats.norm.pdf(x, h_Salmons.mu, h_Salmons.sigma)
```

```python

x = np.linspace(0, 80, 100)

x_test = 45

plt.plot(x, gaussain_mackarels(x), linewidth=2, color='g', label="f(x|Mackerel)")

plt.plot(x, gaussain_salmons(x), linewidth=2, color='r', label="f(x|Salmon)")

plt.plot(x_test, gaussain_mackarels(x_test), 'go')

plt.plot(x_test, gaussain_salmons(x_test), 'ro')

plt.axvline(x_test, -1, 1, linestyle='--', label="x_test = 45")

plt.legend()

plt.show()
```

### if-elif-else

image.png

```python
score = 70            #@param ["47", "50", "60", "70", "80"] {type:"raw", allow-input: true}

if score >= 80:
  print("A")
elif score >= 70:
  print("B")
elif score >= 60:
  print("C")
elif score >= 50:
  print("D")
else:
  print("Fail.")
```

```python
## Nested Block
score = 60            #@param ["47", "50", "60", "70", "80"] {type:"raw", allow-input: true}

if score >= 50:
  print("Pass")
  if score >= 80:
    print("A")
  elif score >= 70:
    print("B")
  elif score >= 60:
    print("C")
  elif score >= 50:
    print("D")
else:
  print("Fail.")
```

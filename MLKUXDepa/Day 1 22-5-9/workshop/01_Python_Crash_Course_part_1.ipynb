### if-elif-else

image.png

```python
score = 70            #@param ["47", "50", "60", "70", "80"] {type:"raw", allow-input: true}

if score >= 80:
  print("A")
elif score >= 70:
  print("B")
elif score >= 60:
  print("C")
elif score >= 50:
  print("D")
else:
  print("Fail.")
```

```python
## Nested Block
score = 60            #@param ["47", "50", "60", "70", "80"] {type:"raw", allow-input: true}

if score >= 50:
  print("Pass")
  if score >= 80:
    print("A")
  elif score >= 70:
    print("B")
  elif score >= 60:
    print("C")
  elif score >= 50:
    print("D")
else:
  print("Fail.")
```

## การรับ Input

* ใช้คำสั่ง `input()` เพื่อรับข้อมูลเข้า
* ข้อมูลเข้าจะเป็นสตริงเสมอ
* ถ้าต้องการข้อมูลเข้าเป็นจำนวนเต็ม ให้ใช้ `int()` เพื่อแปลงสตริงของตัวเลขให้เป็นจำนวนเต็ม
* ถ้าต้องการข้อมูลเข้าเป็นจำนวนจริง ให้ใช้ `float()` เพื่อแปลงสตริงของตัวเลขให้เป็นจำนวนจริง

```python
score = input()
score = int(score)
score = score + 1

if score > 50:
  print("Pass.")
else:
  print("Fail.")
```

```python
score = input()
score = float(score)
score = score * 1.1

if score > 50:
  print("Pass.")
else:
  print("Fail.")
```

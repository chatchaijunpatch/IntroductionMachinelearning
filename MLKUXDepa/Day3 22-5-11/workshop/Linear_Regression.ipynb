# Linear Regression

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Dataset/03-cities.csv',encoding='tis620')
df
```

```python
X = df[['latitude','longitude']]
y=df['temperature']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
lm.fit(X_train,y_train)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
lm.predict([[15,3],[7,105]])
```

```python
lm.score(X_test,y_test)
```

```python
y_train.mean() - y_train # error Rsquen คือสัดส่วนความไม่แน่นอนของความแปรปรวน
```

# Poly Features


```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_transformer=PolynomialFeatures(degree=3,include_bias=False)
```

```python
x_train_transformed=poly_transformer.fit_transform(X_train)
```

```python
pd.DataFrame(x_train_transformed)
```

```python
lm.fit(x_train_transformed,y_train)
```

```python
lm.score(poly_transformer.transform(X_test),y_test)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error
```

```python
y_predicted=lm.predict(poly_transformer.transform(X_test))
```

```python
mean_absolute_error(y_test,y_predicted)
```

```python
mean_absolute_percentage_error(y_test,y_predicted)
```

```python
mean_squared_error(y_test,y_predicted)**0.5 # RMSE
```

# Stochastic Gradient Descent

```python
#ควรมี 5 หมื่นตัวข้อมูลจำนวนแถว
from sklearn.linear_model import SGDRegressor
```

```python
sgd_reg = SGDRegressor(max_iter=100000,tol=1e-5,penalty=None,eta0=0.01,n_iter_no_change=100)
```

```python
sgd_reg.fit(X_train,y_train)
```

```python
sgd_reg.intercept_
```

```python
sgd_reg.coef_
```

```python
sgd_reg.score(X_test,y_test)
```

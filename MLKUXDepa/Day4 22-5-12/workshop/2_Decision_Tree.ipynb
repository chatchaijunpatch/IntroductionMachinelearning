```python
import pandas as pd
```

```python
df1= pd.read_csv('/content/drive/MyDrive/Dataset/01-census-income.csv')
df1.columns
```

```python
df2=pd.read_csv('/content/drive/MyDrive/Dataset/02-future-census.csv')
```

```python
cols=['age','edu num','marital status','sex','captial-gain','capital-loss','hours-per-week']
```

```python
X_train=df1[cols]
X_train
```

```python
y_train=df1[['label']]
```

```python
y_train
```

```python
X_test=df2[cols]
y_test=df2[['label']]
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
```

```python
transformer = make_column_transformer((OneHotEncoder(),['marital status','sex']),remainder='passthrough') #remainder คอลัมไม่กำหนดจะทำไรกับมันดี อันดังกล่าวไม่ทิ้ง
```

```python
X_train_transformed = transformer.fit_transform(X_train)
```

```python
transformer.get_feature_names_out()
```

```python
X_test_transformed=transformer.transform(X_test)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree =DecisionTreeClassifier()
```

```python
dtree.fit(X_train_transformed, y_train)
```

```python
dtree.score(X_test_transformed,y_test)
```

```python
dtree.score(X_train_transformed,y_train)
```

```python
dtree.tree_
```

```python
dtree2=DecisionTreeClassifier(min_samples_split=30,max_depth=10,min_samples_leaf=100)
```

```python
dtree2.fit(X_train_transformed,y_train)
```

```python
dtree2.score(X_train_transformed,y_train)
```

```python
dtree2.score(X_test_transformed,y_test)
```

```python
dtree.feature_importances_
```

# Regressor

```python
df3=pd.read_csv('/content/drive/MyDrive/Dataset/03-cities.csv',encoding='tis620')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df3[['latitude','longitude']]
y=df3[['temperature']]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt_reg = DecisionTreeRegressor(max_depth=5)
dt_reg.fit(X_train,y_train)
```

```python
dt_reg.score(X_train,y_train)
```

```python
dt_reg.score(X_test,y_test)
```
